Imports

```python
from rdflib import Graph, URIRef, Literal 
from rdflib.namespace import RDF, XSD
```

Load the RDF data 

```python
g = Graph() 
g.parse("data/fokg-sw-train-2024.nt", format="nt")
```

List all subjects, predicates and objects

```python
query = """
SELECT ?subject ?predicate ?object
WHERE {
    ?subject ?predicate ?object .
}
"""

results = g.query(query)

for row in results:
    print(f"Subject: {row.subject}, Predicate: {row.predicate}, Object: {row.object}")
```

Count the number of triples

```python
query = """
SELECT (COUNT(*) as ?totalTriples)
WHERE {
    ?subject ?predicate ?object .
}
"""

results = g.query(query)

for row in results: 
    print(f"Total Triples: {row.totalTriples}")
```

List all unique subjects

```python
query = """
SELECT DISTINCT ?subject
WHERE {
    ?subject ?predicate ?object .
}
"""

results = g.query(query)

for row in results: 
    print(f"Subject: {row.subject}")
```

Find Triples with not rdf:Statement Type

```python
query = """
SELECT ?subject ?predicate ?object
WHERE {
    ?subject ?predicate ?object .
    FILTER(?predicate != <http://www.w3.org/1999/02/22-rdf-syntax-ns#type> || ?object != <http://www.w3.org/1999/02/22-rdf-syntax-ns#Statement>)
}
"""

results = g.query(query)

print("Triples with not rdf:Statement Type:")
for row in results:
    print(f"Subject: {row[0]}, Predicate: {row[1]}, Object: {row[2]}")
```

Analyze truth values

```python
query = """
SELECT ?subject ?truthValue
WHERE {
    ?subject <http://swc2017.aksw.org/hasTruthValue> ?truthValue .
}
"""

results = g.query(query)

for row in results:
    print(f"Subject: {row.subject}, Truth Value: {row.truthValue}")
```
